```python
import sympy
from sympy import symbols, sin, cos, solve, diff, integrate
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Definindo Símbolos
x, y, z = symbols('x y z')
```

```python
# 2. Criando uma Expressão
expressao = x**2 + 2*x*y + y**2
```

```python
# 3. Simplificando a Expressão
expressao_simplificada = sympy.simplify(expressao)
print(f"Expressão Simplificada: {expressao_simplificada}")  # Output: (x + y)**2
```

```python
# 4. Resolvendo uma Equação
equacao = x**2 - 4
solucoes = solve(equacao, x)
print(f"Soluções da Equação: {solucoes}")  # Output: [-2, 2]
```

```python
# 5. Calculando Derivada
funcao = sin(x) + cos(x)
derivada = diff(funcao, x)
print(f"Derivada da Função: {derivada}")  # Output: -sin(x) + cos(x)
```

```python
# 6. Calculando Integral
integral = integrate(funcao, x)
print(f"Integral da Função: {integral}") # Output: sin(x) - cos(x)
```

```python
# 7. Renderização em LaTeX
latex_string = sympy.latex(expressao)
print(f"Expressão em LaTeX: {latex_string}")  # Output: x^{2} + 2 x y + y^{2}
```

```python
# Renderizar a expressão em LaTeX
sympy.init_printing()
print(f"Expressão Renderizada:")
expressao
```

```python
# 8. Plotagem de Gráfico (usando Matplotlib)
x_vals = np.linspace(-5, 5, 100)  # Cria 100 valores de x entre -5 e 5
y_vals = np.sin(x_vals)           # Calcula o seno de cada valor de x

plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Gráfico da Função Seno')
plt.grid(True)
plt.show()
```

```python
#9.  Substituição de Valores
expressao = x**2 + y
expressao_substituida = expressao.subs({x: 2, y: 3})
print(f"Expressão com valores substituídos: {expressao_substituida}") #Output: 7
```
